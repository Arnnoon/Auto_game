```python
import pyautogui
import sys
import time
import random
import keyboard
import win32api, win32con

# Centralized settings for positions and RGB values
SETTINGS = {
    'autoRTA': {
        'start_normal': {
            'check': {'pos': (1545, 478), 'color': (255, 250, 201)},
            'click': {'area': ((1414, 441), (1738, 539))}
        },
        'start_special': {
            'check': {'pos': (1389, 877), 'color': (255, 250, 201)},
            'click': {'area': ((1264, 843), (1593, 932))}
        },
        'out_of_ticket': {
            'check': {'pos': (934, 632), 'color': (248, 233, 173)},
            'click': {'area': ((844, 599), (1032, 661))}
        },
        'give_up': {
            'check': {'pos': (934, 633), 'color': (248, 233, 173)},
            'click': {'area': ((841, 602), (1033, 667))}
        },
        'autoplay': {
            'check': {'pos': (245, 961), 'color': (255, 255, 255)},
            'click': {'area': ((219, 933), (280, 991))}
        },
        'match_finish_normal': {
            'check': {'pos': (361, 773), 'color': (201, 103, 18)},
            'click': {'area': ((1072, 599), (1581, 840))}
        },
        'match_finish_special': {
            'check': {'pos': (361, 768), 'color': (237, 179, 82)},
            'click': {'area': ((1072, 599), (1581, 840))}
        },
    },
    'autoFarmGrindStone': {
        'start_farm_grid': {
            'check': {'pos': (1596, 922), 'color': (252, 252, 252)},
            'click': {'area': ((1465, 901), (1756, 996))}
        },
        'start_farm_gem': {
            'check': {'pos': (1485, 747), 'color': (252, 252, 252)},
            'click': {'area': ((1374, 726), (1656, 822))}
        },
        'round_complete': {
            'check': {'pos': (1011, 886), 'color': (254, 254, 254)},
            'click': [
                {'area': ((1487, 898), (1723, 930))},   # Confirm complete
                {'area': ((1303, 885), (1491, 940))},   # Sell
                {'area': ((682, 601), (875, 666))},     # Confirm sell
                {'area': ((874, 864), (1116, 917))}     # go next round
            ]
        },
        'nothing_sell': {
            'check': {'pos': (1078, 423), 'color': (207, 178, 118)},
            'click': [
                {'area': ((844, 600), (1026, 659))},
                {'area': ((1542, 889), (1723, 935))}
            ]
        },
        'high_grade_gem_sell': {
            'check': {'pos': (716, 548), 'color': (121, 94, 73)},
            'click': [
                {'area': ((707, 537), (727, 551))},
                {'area': ((676, 658), (883, 730))}
            ]
        },
        'out_of_energy': {
            'check': {'pos': (1109, 406), 'color': (226, 197, 130)}
            # 'click': {'area': ((685, 660), (871, 718))}
        }
    },
    'autoFarmRune': {
        'start_farm_rune': {
            'check': {'pos': (1579, 756), 'color': (252, 252, 252)},
            'click': {'area': ((1454, 736), (1731, 826))}
        },
        'round_complete': {
            'check': {'pos': (1011, 886), 'color': (254, 254, 254)},
            'click': [
                {'area': ((1487, 898), (1723, 930))},   # Confirm complete
                {'area': ((1303, 885), (1491, 940))},   # Sell
                {'area': ((682, 601), (875, 666))},     # Confirm sell
                {'area': ((874, 864), (1116, 917))}     # go next round
            ]
        },
        'nothing_sell': {
            'check': {'pos': (1075, 420), 'color': (207, 178, 118)},
            'click': [
                {'area': ((840, 595), (1025, 655))},
                {'area': ((1535, 885), (1720, 930))}
            ]
        },
        'high_grade_rune_sell': {
            'check': {'pos': (1214, 326), 'color': (226, 197, 130)},
            'click': [
                {'area': ((705, 535), (725, 550))},
                {'area': ((675, 655), (880, 725))}
            ]
        },
        'out_of_energy': {
            'check': {'pos': (1105, 405), 'color': (226, 197, 130)}
        }
    }
}

# Click and random click functions
def click(x, y):
    original_x, original_y = win32api.GetCursorPos()
    win32api.SetCursorPos((x, y))
    win32api.mouse_event(win32con.MOUSEEVENTF_LEFTDOWN, 0, 0)
    time.sleep(0.01)
    win32api.mouse_event(win32con.MOUSEEVENTF_LEFTUP, 0, 0)
    win32api.SetCursorPos((original_x, original_y))

def dragclick(x1, y1, x2, y2):
    original_x, original_y = win32api.GetCursorPos()
    win32api.SetCursorPos((x1, y1))
    win32api.mouse_event(win32con.MOUSEEVENTF_LEFTDOWN, 0, 0)
    time.sleep(0.1)
    win32api.SetCursorPos((x2, y2))
    time.sleep(0.1)
    win32api.mouse_event(win32con.MOUSEEVENTF_LEFTUP, 0, 0)
    win32api.SetCursorPos((original_x, original_y))

def randclick(area, wait=1):
    """Randomly click within a specified area in SETTINGS."""
    top_left, bottom_right = area
    x1, y1 = top_left
    x2, y2 = bottom_right
    center_x = (x1 + x2) / 2
    center_y = (y1 + y2) / 2
    max_radius_x = center_x - x1
    max_radius_y = center_y - y1
    x_scale = random.uniform(-1, 1)
    x_scale = x_scale * abs(x_scale)
    y_scale = random.uniform(-1, 1)
    y_scale = y_scale * abs(x_scale)
    xx = int(center_x + (x_scale * max_radius_x))
    yy = int(center_y + (y_scale * max_radius_y))
    click(xx, yy)
    randtimesleep(wait)
    # print(f"Clicked at ({xx}, {yy}) in {area}")

# Get pixel color function
def get_pixelcolor(position):
    return pyautogui.screenshot(region = (0,0,1919,1079)).getpixel(position)

# Check color function
def checkcolor(position, expected_color, tolerance=5):
    """Check if the color at the given position matches the expected color within a tolerance."""
    time.sleep(1)
    if keyboard.is_pressed('left alt'): sys.exit()
    r, g, b = get_pixelcolor(position)
    return (abs(r - expected_color[0]) <= tolerance and
            abs(g - expected_color[1]) <= tolerance and
            abs(b - expected_color[2]) <= tolerance)

# Random sleep function
def randtimesleep(min_time=1, max_time=1.5):
    """Sleep for a random duration between min_time and max_time."""
    sleep_time = random.uniform(min_time, max_time)
    time.sleep(sleep_time)

# Exit game function
def exitgame():
    click(1860, 16)
    time.sleep(0.5)
    click(1011, 575)

# continuous click
def continuousclick():
    while keyboard.is_pressed('left alt') == False:
        if keyboard.is_pressed('s') == True:
            x, y = pyautogui.position()
            click(x, y)
            time.sleep(0.01)

# Print position and color value (r, g, b) of a given position
def checkPosition(x, y):
    while not keyboard.is_pressed('left alt'):
        if keyboard.is_pressed('s'):
            r, g, b = get_pixelcolor((x, y))
            print(f"# x,y ({x}, {y})   r,g,b ({r}, {g}, {b})\t\tif checkcolor(({x}, {y}), ({r}, {g}, {b})):")
            time.sleep(0.1)
    return 1

# Print position and color value (r, g, b) of the mouse position and define clickable areas
def checkClick():
    x1 = y1 = x2 = y2 = -1
    while not keyboard.is_pressed('left alt'):
        if keyboard.is_pressed('s'):
            x, y = pyautogui.position()
            r, g, b = get_pixelcolor((x, y))
            print(f"# x,y ({x}, {y})   r,g,b ({r}, {g}, {b})\twhile not checkcolor(({x}, {y}), ({r}, {g}, {b})):")
            print(f"""'': {{'pos': ({x}, {y}), 'color': ({r}, {g}, {b})}}""")
            time.sleep(0.1)
        elif keyboard.is_pressed('a'):
            x1, y1 = pyautogui.position()
        elif keyboard.is_pressed('d'):
            x2, y2 = pyautogui.position()

        if x1 != -1 and y1 != -1 and x2 != -1 and y2 != -1:
            print(f"# x1,y1 ({x1}, {y1})  x2,y2 ({x2}, {y2})\trandclick(({x1}, {y1}), ({x2}, {y2}))")
            print(f"""'': {{'area': (({x1}, {y1}), ({x2}, {y2}))}}""")
            x1 = y1 = x2 = y2 = -1
            time.sleep(0.1)
    return 1


# Main automation function
def autoRTA(times, type='normal'):
    round = 0
    settings = SETTINGS['autoRTA']
    print("autoRTA will start in 3 sec")
    time.sleep(3)
    print("autoRTA is starting")

    if times == 0:
        times = 999999

    start_time = time.strftime("%H:%M:%S", time.localtime())
    finish_time = time.strftime("%H:%M:%S", time.localtime())

    while not keyboard.is_pressed('left alt'):
        if round >= times:
            break

        # Check the start button color
        if checkcolor(settings['start_'+type]['check']['pos'], settings['start_'+type]['check']['color']):
            round += 1
            print(f"Round {round} starting")
            # Click in the start button area
            randclick(settings['start_'+type]['click']['area'])
            
            # Check if out of ticket notification is present
            if checkcolor(settings['out_of_ticket']['check']['pos'], settings['out_of_ticket']['check']['color']):
                round -= 1
                print("\tOut of ticket")
                randclick(settings['out_of_ticket']['click']['area'])
                break

            while not keyboard.is_pressed('left alt'):
                # Handle give up notification
                if checkcolor(settings['give_up']['check']['pos'], settings['give_up']['check']['color']):
                    print("\tGive up notification")
                    randclick(settings['give_up']['click']['area'])
                    break
                
                # Check for autoplay button notification
                if checkcolor(settings['autoplay']['check']['pos'], settings['autoplay']['check']['color']):
                    print("\tAutoplay found")
                    randclick(settings['autoplay']['click']['area'])
                
                    while not keyboard.is_pressed('left alt'):
                        # Check for match finish
                        if checkcolor(settings['match_finish_'+type]['check']['pos'], settings['match_finish_'+type]['check']['color']):
                            print("\tMatch finish detected")
                            randclick(settings['match_finish_'+type]['click']['area'])
                            break
                    break
        randtimesleep(1)

    finish_time = time.strftime("%H:%M:%S", time.localtime())
    print(f"\n{round} rounds")
    print("Start time:", start_time)
    print("Finish time:", finish_time)
    exitgame()
    return 0
    
def autoFarmGrindStone():
    round = 0
    settings = SETTINGS['autoFarmGrindStone']
    print("Auto Grind Stone farm until energy is out will start when press 's'")
    
    while not keyboard.is_pressed('left alt'):
        if keyboard.is_pressed('s'):
            while not keyboard.is_pressed('left alt'):
                # Start farming
                round += 1
                start_time = time.strftime("%H:%M:%S", time.localtime())
                finish_time = time.strftime("%H:%M:%S", time.localtime())
                print(f"\tStarts round {round}")
                print(f"\tStart round time: {start_time}")

                if checkcolor(settings['start_farm_grid']['check']['pos'], settings['start_farm_grid']['check']['color']):
                    randclick(settings['start_farm_grid']['click']['area'])
                    print('\tMode: Grid')
                else:
                    randclick(settings['start_farm_gem']['click']['area'])
                    print('\tMode: Gem')
                
                # Check if out of energy
                if checkcolor(settings['out_of_energy']['check']['pos'], settings['out_of_energy']['check']['color']):
                    print("Exit: out of energy")
                    break

                while not keyboard.is_pressed('left alt'):
                    # Check for round completion
                    if checkcolor(settings['round_complete']['check']['pos'], settings['round_complete']['check']['color']):
                        for c in settings['round_complete']['click'][:-2]:
                            randclick(c['area'])

                        # Handle selling
                        if checkcolor(settings['nothing_sell']['check']['pos'], settings['nothing_sell']['check']['color']):
                            for c in settings['nothing_sell']['click']:
                                randclick(c['area'])
                            print("\tNothing to sell")
                        else:
                            for c in settings['round_complete']['click'][2:-1]:
                                randclick(c['area'])
                            print("\tSomething to sell")
                            
                            # Check for high-grade gem selling
                            if checkcolor(settings['high_grade_gem_sell']['check']['pos'], settings['high_grade_gem_sell']['check']['color']):
                                for c in settings['high_grade_gem_sell']['click']:
                                    randclick(c['area'])
                                print("\tSell high-grade gem")

                        randclick(settings['round_complete']['click'][-1]['area'])
                        break
                    time.sleep(10)
            break
    
    finish_time = time.strftime("%H:%M:%S", time.localtime())
    print(f"\n{round} rounds completed")
    print(f"Start time: {start_time}")
    print(f"Finish time: {finish_time}")
    return 0

def autoFarmRune():
    round = 0
    settings = SETTINGS['autoFarmRune']
    print("Auto Rune farm until energy is out will start when press 's'")
    
    while not keyboard.is_pressed('left alt'):
        if keyboard.is_pressed('s'):
            while not keyboard.is_pressed('left alt'):
                # Start farming
                round += 1
                start_time = time.strftime("%H:%M:%S", time.localtime())
                finish_time = time.strftime("%H:%M:%S", time.localtime())
                print(f"\tStarts round {round}")
                print(f"\tStart round time: {start_time}")

                if checkcolor(settings['start_farm_rune']['check']['pos'], settings['start_farm_rune']['check']['color']):
                    randclick(settings['start_farm_rune']['click']['area'])
                
                # Check if out of energy
                if checkcolor(settings['out_of_energy']['check']['pos'], settings['out_of_energy']['check']['color']):
                    print("Exit: out of energy")
                    break

                while not keyboard.is_pressed('left alt'):
                    # Check for round completion
                    if checkcolor(settings['round_complete']['check']['pos'], settings['round_complete']['check']['color']):
                        for c in settings['round_complete']['click'][:-2]:
                            randclick(c['area'])

                        # Handle selling
                        if checkcolor(settings['nothing_sell']['check']['pos'], settings['nothing_sell']['check']['color']):
                            for c in settings['nothing_sell']['click']:
                                randclick(c['area'])
                            print("\tNothing to sell")
                        else:
                            for c in settings['round_complete']['click'][2:-1]:
                                randclick(c['area'])
                            print("\tSomething to sell")
                            
                            # Check for high-grade gem selling
                            if checkcolor(settings['high_grade_rune_sell']['check']['pos'], settings['high_grade_rune_sell']['check']['color']):
                                for c in settings['high_grade_rune_sell']['click']:
                                    randclick(c['area'])
                                print("\tSell high-grade gem")

                        randclick(settings['round_complete']['click'][-1]['area'])
                        break
                    time.sleep(10)
            break
    
    finish_time = time.strftime("%H:%M:%S", time.localtime())
    print(f"\n{round} rounds completed")
    print(f"Start time: {start_time}")
    print(f"Finish time: {finish_time}")
```
